```python
import pandas as pd
import sfacts as sf
```

```python
prefix = 'data/sfacts_simulate-model_simplest_simulation-n50-g250-s10-pi40-mu100-eps10-seed0'
```

```python
sim = sf.World.load(f'{prefix}.world.nc')
```

```python
mix0 = sf.World.load(f'{prefix}.metagenotype-n50-g250.fit-mixtureS.world.nc')
mix = mix0.collapse_strains(0.05)
fac = sf.World.load(f'{prefix}.metagenotype-n50-g250.fit-sfacts44-s15-seed0.world.nc')
fin = sf.World.load(f'{prefix}.metagenotype-n50-g250.fit-sfinder-s10-seed0.world.nc')
```

```python
sf.plot_community(sim, row_linkage_func=lambda w: w.genotypes.linkage(), col_linkage_func=lambda w: sim.communities.linkage('sample'))
```

```python
sf.plot_community(mix, row_linkage_func=lambda w: w.genotypes.linkage(), col_linkage_func=lambda w: sim.communities.linkage('sample'))
```

```python
w = sf.World.concat(dict(fac=fac, mix=mix, mix0=mix0, sim=sim, fin=fin), 'strain', rename_coords=True)

sf.plot_genotype(w, scaley=0.2, )
```

```python
sf.plot_community(fac, row_linkage_func=lambda w: w.genotypes.linkage(), col_linkage_func=lambda w: sim.communities.linkage('sample'))
```

```python
sf.plot_community(fin, row_linkage_func=lambda w: w.genotypes.linkage(), col_linkage_func=lambda w: sim.communities.linkage('sample'))
```
